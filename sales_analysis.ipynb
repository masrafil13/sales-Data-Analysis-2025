### **Objective**
Analyze and visualize retail sales data for 2025 using Python, Excel, and Tableau to identify trends and growth insights.


```python

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_excel("cleaned_sales_data_2025.xlsx")

# Display first few rows
df.head()
```

```python

# Check data information
df.info()

# Summary statistics
df.describe()
```

```python

# Add month column for analysis
df['Month'] = df['Date'].dt.strftime('%b')

# Total sales and profit
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python

# Sales by region
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python

# Monthly trend
monthly_sales = df.groupby('Month')['Sales'].sum().reindex(
    ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
)
monthly_sales.plot(kind='line', marker='o', title="Monthly Sales Trend (2025)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python

# Top products
top_products = df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(5)
sns.barplot(x=top_products.values, y=top_products.index)
plt.title("Top 5 Products by Sales")
plt.xlabel("Sales")
plt.ylabel("Product")
plt.show()
```
